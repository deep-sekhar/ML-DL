```python
import numpy as np 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Movie.csv', header = 0)
```

```python
df.head()
```

```python
df.info()
```

### missing values

```python
df['Time_taken'].mean()
```

```python
df['Time_taken'].fillna(value = df['Time_taken'].mean(), inplace = True)
```

```python
df.info()
```

### dummy variable

```python
df.head()
```

```python
df = pd.get_dummies(df, columns=['3D_available', 'Genre'], drop_first=True)
```

```python
df.head()
```

### X-y split

```python
X = df.loc[:, df.columns!='Collection']
y = df['Collection']
```

```python
X.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
X_train.head()
```

```python
X_train.shape
```

### decision tree

#### dont exceed depth beyond 5

```python
from sklearn import tree
regtree = tree.DecisionTreeRegressor(max_depth=3)
```

```python
regtree.fit(X_train, y_train)
```

```python
y_train_pred = regtree.predict(X_train)
y_test_pred = regtree.predict(X_test)
```

### performance

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
mean_squared_error(y_train, y_train_pred)
```

```python
r2_score(y_train, y_train_pred)
```

```python
r2_score(y_test, y_test_pred)
```

### plot decision tree

```python
dot_data = tree.export_graphviz(regtree, out_file=None)
```

```python
from IPython.display import Image
```

```python
import pydotplus
```

```python
graph = pydotplus.graph_from_dot_data(dot_data)
Image(graph.create_png())
```

## Tree pruning and control tree growth

python does not support post pruning

### 1.max no of levels

```python
regtree1 = tree.DecisionTreeRegressor(max_depth=3)
regtree1.fit(X_train, y_train)
dot_data = tree.export_graphviz(regtree1, out_file=None,feature_names=X_train.columns, filled = True)
graph = pydotplus.graph_from_dot_data(dot_data)
Image(graph.create_png())
```

### 2.Min observations in internal node

```python
regtree2 = tree.DecisionTreeRegressor(max_depth=3, min_samples_split=3)
regtree2.fit(X_train, y_train)
dot_data = tree.export_graphviz(regtree2, out_file=None,feature_names=X_train.columns, filled = True)
graph2 = pydotplus.graph_from_dot_data(dot_data)
Image(graph2.create_png())
```

### 3.Min obs at leaf nodes

```python
regtree3 = tree.DecisionTreeRegressor(max_depth=3, min_samples_leaf=3)
regtree3.fit(X_train, y_train)
dot_data = tree.export_graphviz(regtree3, out_file=None,feature_names=X_train.columns, filled = True)
graph3 = pydotplus.graph_from_dot_data(dot_data)
Image(graph3.create_png())
```
